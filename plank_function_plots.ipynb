```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def blackbody(wave, T):
    front = (2 * 6.62607004 * (10 ** -34) * 299792458**2)/(wave ** 5)
    back = 1/(np.exp((6.62607004 * (10 ** -34) * 299792458)/(wave * 1.38064852 * (10 ** -23) * T)) - 1)
    return front * back
```

```python
wavelengths = np.linspace(3.5*10**-7, 1.6*10**-6, 10**7)
```

```python
x = wavelengths
T = [300, 1000, 3000, 5000]
y = []
for i in T:
    y.append(blackbody(wavelengths, i))

x_nm = x * 10**9
```

```python
for yi, ti in zip(y, T):
    plt.plot(x_nm, yi, label = "{}K".format(ti))
    plt.xlabel('$\lambda$ (nm)')
    plt.ylabel('Wsr$^{-1}$m$^{-3}$')
    plt.title("Plank Function at {}K".format(ti))
    plt.show()
```
